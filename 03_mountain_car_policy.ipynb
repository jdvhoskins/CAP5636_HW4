# Cross-Entropy Mountain Car Policy

Apply the cross-entropy method to mountain car. Explain how many episodes are needed to learn a good policy. Explain which reward you use (original, modified).

```python
import gym
import tensorflow as tf
import numpy as np
from collections import deque


%matplotlib inline
import matplotlib.pyplot as plt

import time
env = gym.make("MountainCar-v0")
```

# Hyperparameters

While testing the performance of the QNN, I altered some of the parameters listed below. These included action_size, train_episodes, decay_rate, first_layer_size, and memory_size.

action_size - I changed this value to 2 ("left" & "right" only) and I modified the qnn and pretraining code accordingly to not use "no action". This degraded the performance of the qnn, and the value was reverted to 3.

train_episodes - I increased this from 501 to 1001 in order to help identify and quantify overtraining. After about 300 episodes, the performance of the agent would stop improving and would start regressing. This was clearly due to over fitting based on the modified reward. Changes were made to the modified reward function to minimize this effect. This value was then reverted to 501.

decay_rate - I changed this value to 0.001 and 0.005 to allow for longer periods of exploration before eploitation took over. There was no noticable improvement in the final performance of the agent as a result of lowering the decay_rate, and the value was reverted to 0.01.

first_layer_size - I changed this value from 64 to 128. The qnn yielded positive results for the agent using slightly fewer training episodes. This change was left in place.

memory_size - I changed this value to 10000 and 30000 representing a history of 50 and 150 episodes respectively. Decreasing the episode history had a drastic, negative effect on the qnn, while increasing the episode history had a slight, positive effect instead. This value was left at 30000.



```python
# Environment parameters
state_size = 2
action_size = 3

# Training parameters
train_episodes = 501          # number of episodes to train
max_steps = 200                # max number of step in an episode
gamma = 0.99                   # future reward discount

# Exploration parameters
epsilon_start = 1.0            # exploration probability at start
epsilon_stop = 0.01            # minimum exploration probability 
decay_rate = 0.01            # exponential decay rate for exploration prob

# Q neural network parameters
first_layer_size = 128         # number of neurons in first hidden layer
second_layer_size = 64         # number of neurons in second hidden layer
learning_rate = 0.0001         # learning rate

# Memory parameters
memory_size = 30000             # memory capacity
batch_size = 100                # experience mini-batch size
pretrain_length = memory_size   # number experiences to pretrain the memory
```

# Modified Reward

The modified reward function was changed to include a small reward for negative displacement of the car. This is to encourage some backward motion which is required to ultimately propel the car up the hill on the right. To keep the qnn from excessively wanting to go left, the modified reward for positions left of -0.5 are reduced by a factor of 10.

Originally, a small bonus was added to the modified reward if the car made it to the finish line. This small reward helped in the early training episodes, but it proved ineffective after about 300 episodes. The agent was learning to maximize the modified reward which, due to the small size of the bonus reward, is done by getting close to the finish line and staying there as long as possible. To fix this, the total reward for crossing the finish line was asigned a static value later on.

```python
def compute_modified_reward(next_state):
    modified_reward = np.square(next_state[0] + 0.5)
    
    if next_state[0] >= 0.5: 
        modified_reward += 1.0

    if next_state[0] < -0.5: 
        modified_reward /= 10.0 
    
    return modified_reward


class Memory():
    def __init__(self, max_size = 1000):
        self.buffer = deque(maxlen = max_size)
    
    def add(self, experience):
        self.buffer.append(experience)
            
    def sample(self, batch_size):
        random_index_list = np.random.choice(np.arange(len(self.buffer)), 
                               size = batch_size, 
                               replace = False)
        return [self.buffer[index] for index in random_index_list]
```

```python
# Create empty queue
memory = Memory(max_size = memory_size)

# Make a bunch of random actions and store the experiences
state = env.reset()

for _ in range(pretrain_length):
    # Take a random action
    action = env.action_space.sample()
    #action = 0 if np.random.rand() < 0.5 else 2
    next_state, _, done, _ = env.step(action)
    modified_reward = compute_modified_reward(next_state)
    
    memory.add((state, action, modified_reward, next_state, done))
    
    if done:
        # Start new episode
        state = env.reset()
    else:
        # Go to next state
        state = next_state
```

```python
# first hidden layer 128 neurons
# second hidden layer 64 neurons

class QNN:
    def __init__(self, 
                 state_size = state_size, 
                 action_size = action_size, 
                 first_layer_size = 128,
                 second_layer_size = 64,
                 learning_rate = 0.01, 
                 name = 'qnn'):
        
        # state inputs to the Q-network
        with tf.variable_scope(name):
    
            # Prediction part of the QNN
            # computes q(s, a) for given state s and all actions a
            # so best action can be determined as argmax_a q(s, a)
    
            # Input layer, state s is input
            self.state = tf.placeholder(tf.float32, 
                                        [None, state_size], 
                                        name = 'state')
            
            # First hidden layer, ReLU activation
            self.first_layer = tf.contrib.layers.fully_connected(self.state, 
                                                                 first_layer_size)
            # Second hidden layer, ReLU activation
            self.second_layer = tf.contrib.layers.fully_connected(self.first_layer, 
                                                                  second_layer_size)

            # Output layer, linear activation, q_vector(s, a) is output
            self.q_vector = tf.contrib.layers.fully_connected(self.second_layer,
                                                              action_size,
                                                              activation_fn = None)
    
            # Training part of the Q-network
            # uses observed transition (s, a, r, s') to update the weights of the network
    
            # Action a
            self.action = tf.placeholder(tf.int32, 
                                         [None], 
                                         name = 'action')
            # One-hot encoded action a 
            #
            # encoded_action_vector = [1, 0] if action a = 0
            # encoded_action_vector = [0, 1] if action a = 1
            encoded_action_vector = tf.one_hot(self.action, 
                                               action_size)
    
            # Target Q value for training
            # target_q_value = r + gamma * max_a' q(s', a') if state s' is non-terminal
            # target_q_value = r                            if state s' is terminal
            self.target_q_value = tf.placeholder(tf.float32, 
                                                 [None], 
                                                 name = 'target_q_value')
    
            # Q value for training
            # q_value = q(s, a)
            # It is important that everything is computed using linear-algebraic operations
            # to leverage vectorization etc.
            q_value = tf.reduce_sum(tf.multiply(self.q_vector, encoded_action_vector), 
                                         axis = 1)
            # Loss
            self.loss = tf.reduce_mean(tf.square(self.target_q_value - q_value))
            
            # Optimizer
            self.optimizer = tf.train.AdamOptimizer(learning_rate).minimize(self.loss)
            
    # get best action 
    def get_action(self, state):
        feed_dict = { self.state : np.array([state]) } 
        q_vector = sess.run(self.q_vector, 
                            feed_dict = feed_dict)
        return np.argmax(q_vector)
        
    # train based on a batch of data from 
    def update(self):    
        
        # Sample mini-batch from memory
        batch = memory.sample(batch_size)
        # some Python magic
        state, action, reward, next_state, done = zip(*batch)
    
        state = np.array(state)
        action = np.array(action)
        reward = np.array(reward)
        next_state = np.array(next_state)
        mask = 1.0 - np.array(done)
    
        next_q_vector = sess.run(self.q_vector, 
                                 feed_dict = { 
                                     self.state : next_state 
                                 })
        
        target_q_value = reward + gamma * mask * np.max(next_q_vector, axis = 1)

        sess.run([self.loss, self.optimizer],
                 feed_dict = {
                               self.state          : state,
                               self.target_q_value : target_q_value,
                               self.action         : action
                 })
```

# Define QNN

Also define a function for running the agent using the current qnn.

```python
tf.reset_default_graph()
qnn = QNN(name = 'qnn', 
          first_layer_size = first_layer_size,
          second_layer_size = second_layer_size,
          learning_rate = learning_rate)

def run_policy(env, qnn, max_steps, render):
    state = env.reset()
    max_pos = state[0]
    total_reward = 0
    total_modified_reward = 0
    for step in range(max_steps):
        if render:
            env.render()
            time.sleep(0.01)
        next_state, reward, done, _ = env.step(qnn.get_action(state))
        modified_reward = compute_modified_reward(next_state)
        
        total_reward += reward
        total_modified_reward += modified_reward
        if next_state[0] > 0.5:
            total_modified_reward = 200
        
        if next_state[0] > max_pos:
            max_pos = next_state[0]
        if done: 
            print("Total reward:", total_reward)
            print("Total modified reward:", total_modified_reward)
            print("Max position:", max_pos)
            break
        state = next_state
    
    env.close()
    return [total_reward, total_modified_reward]
```

# Training

Following each episode, a random sellection of 100 history elements is fed to the qnn in order to update the weights.

```python
# saver = tf.train.Saver()

episode_reward_list = []
episode_modified_reward_list = []
print_reward_list = []
print_modified_reward_list = []

with tf.Session() as sess:
    
    # Initialize variables
    sess.run(tf.global_variables_initializer())
    
    #step = 0
    epsilon = epsilon_start
    
    for episode in range(train_episodes):
        total_reward = 0
        total_modified_reward = 0
        state = env.reset()
        
        # Decrease epsilon
        epsilon = epsilon_stop + (epsilon_start - epsilon_stop) * np.exp(-decay_rate * episode) 
        
        for _ in range(max_steps):
            #print("episode, step in episode:", episode, s)
            # Explore or exploit
            if epsilon > np.random.rand():
                # Explore - Select random action
                action = env.action_space.sample()
                #action = 0 if np.random.rand() < 0.5 else 2
            else:
                # Exploit - Compute q_vector using q neural network
                action = qnn.get_action(state)
            
            
            #print(step, epsilon)
            
            # Take epsilon-greedily selected action,
            # obtain next state and reward
            next_state, reward, done, _ = env.step(action)
            modified_reward = compute_modified_reward(next_state)

            if next_state[0] > 0.5:
                total_modified_reward = 200

            # Add transition to memory
            memory.add((state, action, modified_reward, next_state, done))

            #step += 1
            total_reward += reward
            total_modified_reward += modified_reward
            state = next_state
            
            # Update weights of q neural network
            qnn.update()
            
            if done: 
                #print("done!")
                break
        
        # Episode ended because either done or max_steps reached
        
        # Why does this code get executed twice????
        
        episode_reward_list.append((episode, total_reward))
        episode_modified_reward_list.append((episode, total_modified_reward))
        
        #print("Episode just ended:", episode)
        
        if episode % 10 == 0:
            print('\nEpisode: {},'.format(episode),
                  'Epsilon: {:.4f}'.format(epsilon))
            if episode % 100 == 0:
                reward, modified_reward = run_policy(env, qnn, max_steps, True)
            else:
                reward, modified_reward = run_policy(env, qnn, max_steps, False)

            print_reward_list.append((episode, reward))
            print_modified_reward_list.append((episode, modified_reward))
            
        
```

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / N
```

```python
eps, rews = np.array(episode_reward_list).T
smoothed_rews = running_mean(rews, 10)
plt.plot(eps[-len(smoothed_rews):], smoothed_rews)
plt.plot(eps, rews, color='grey', alpha = 0.3)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
```

```python
eps, rews = np.array(print_reward_list).T
eps, mod_rews = np.array(print_modified_reward_list).T
plt.plot(eps, mod_rews)
plt.plot(eps, rews, color='grey')
plt.xlabel('Episode')
plt.ylabel('Rewards')
plt.legend(['Modified Reward','Total Reward'])
```

# Results

The reward needed to be modified in such a way that there would be some indication of progress being made. Using the default reward would have yielded a value of -200 regardles of significant progress towards the finish line. The problem with using the modified reward is that the agent is training for a goal that, while similar, is still distinct from the actual goal. The policy develped may actually be sub-optimal for the real goal, because it is optimal for the modified goal. Generally, this can be avoided by not over-training the agent.
